### (b) Test and Train Data

```python
base_path = "../data/AReM/"
folders = [
    "bending1", "bending2", "cycling", "lying", "sitting", "standing", "walking"
]

def load_and_separate_data():
    test_data = []
    train_data = []
    
    for folder in folders:
        folder_path = os.path.join(base_path, folder)
        
        if folder == "bending1":
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset1.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset2.csv"), sep=',', skiprows=4))
            for i in range(3, 8):
                train_data.append(pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4))
        
        elif folder == "bending2":
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset1.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset2.csv"), sep=',', skiprows=4))
            for i in range(3, 7):
                if i==4:
                    train_data.append(df_bending_4)
                else:
                    train_data.append(pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4))
        
        elif folder == "cycling":
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset1.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset2.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset3.csv"), sep=',', skiprows=4))
            for i in range(4, 16):
                if i==9:
                    train_data.append(df_cycling_9)
                elif i==14:
                    train_data.append(df_cycling_14)
                else:
                    train_data.append(pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4))
        
        else:
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset1.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset2.csv"), sep=',', skiprows=4))
            test_data.append(pd.read_csv(os.path.join(folder_path, "dataset3.csv"), sep=',', skiprows=4))
            for i in range(4, 16):
                train_data.append(pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4))
    
    test_data_combined = pd.concat(test_data, ignore_index=True)
    train_data_combined = pd.concat(train_data, ignore_index=True)
    
    return train_data_combined, test_data_combined

train_data, test_data = load_and_separate_data()
```

### (c) Feature Extraction

#### i. Research

**Types of time-domain features usually used in time series classification are: -**<br><br>
**Statistical Features:**
1. Mean
2. Median
3. Minimum
4. Maximum
5. Range
6. Standard Deviation
7. Variance
8. IQR
<br><br>
**Shape-Based Features:**
1. Skewness
2. Kurtosis
3. Entropy
<br><br>
**Time Series Features:**
1. Orders of autoregressive, integrated and moving average part of an estimated ARIMA model
2. Parameters of the AR part
3. Parameters of the MA part

#### ii. Extraction

```python
base_path = "../data/AReM/"
folders = [
    "bending1", "bending2", "cycling", "lying", "sitting", "standing", "walking"
]
col_names = ['min_', 'max_', 'mean_', 'median_', 'std_dev_', '1st quartile_', '3rd quartile_' ]
col = []
for i in range (1,7):
    for j in col_names:
        col.append(j + str(i))
        
df_2 = pd.DataFrame(columns=col)
def extract_features(d):
    summary = d.describe()
    
    row = []
    for i in range(1, 7):
        row.extend([
            summary.iloc[3, i],  
            summary.iloc[7, i],  
            summary.iloc[1, i],  
            summary.iloc[5, i],  
            summary.iloc[2, i],  
            summary.iloc[4, i],  
            summary.iloc[6, i],  
        ])
    
    return row

for folder in folders:
    folder_path = os.path.join(base_path, folder)

    if folder == "bending1":
        for i in range(1, 8):
            d = pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4)
            row = extract_features(d)
            df_2 = pd.concat([df_2, pd.DataFrame([row], columns=col)], ignore_index=True)
    
    elif folder == "bending2":
        for i in range(1, 7):
            if i==4:
                d = df_bending_4
            else:
                d = pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4)
            row = extract_features(d)
            df_2 = pd.concat([df_2, pd.DataFrame([row], columns=col)], ignore_index=True)

    elif folder == "cycling":
        for i in range(1, 16):
            if i==9:
                d = df_cycling_9
            elif i ==14:
                d = df_cycling_14
            else:
                d = pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4)
            row = extract_features(d)
            df_2 = pd.concat([df_2, pd.DataFrame([row], columns=col)], ignore_index=True)

    else:
        for i in range(1, 16):
            d = pd.read_csv(os.path.join(folder_path, f"dataset{i}.csv"), sep=',', skiprows=4)
            row = extract_features(d)
            df_2 = pd.concat([df_2, pd.DataFrame([row], columns=col)], ignore_index=True)

df_2.insert(0, "Instance", range(1, len(df_2) + 1))

df_2
```

#### iii. Standard Deviation

```python
df_summary = df_2.loc[:, df_2.columns!="Instance"].describe()
std_dev = df_summary.loc[['std']].T
std_dev
```

```python
ci = pd.DataFrame()
for i in range(len(df_2.columns)):
    data = (df_2.iloc[:, i],)
    boot = bootstrap(data, np.std, confidence_level = 0.9, random_state = 1, method = 'percentile')
    vals = pd.DataFrame({'Feature': df_2.columns[i], 'Lower Bound': boot.confidence_interval.low, 'Upper Bound': boot.confidence_interval.high}, index=[i+1])
    ci = pd.concat([ci, vals], ignore_index = True)

ci = ci.iloc[1:].reset_index(drop=True)
ci
```

#### iv. Select Features

The 3 most important time-domain features that I'll select are:<br>
1. Mean
2. Median
3. Standard deviation

**MEAN: - <br>**
Mean's confidence interval gives us the range in which the true average value of the feature lies.<br>
As we can see in the table, the lower bound and upper bound make a quite tight range. <br>**Tighter CI means consistent and predictable behavior of the feature values.**
<br><br>
**MEDIAN: - <br>**
Median's confidence interval gives us the range where the central value of the dataset is expected to be.<br>
As we can see in the table, the lower bound and upper bound values are closer to each other. <br>**Small CI means more stable feature values.**
<br><br>
**Standard Deviation: - <br>**
Standard Deviation's confidence interval shows how unceratinity in feature values.<br>
As we can see in the table, the difference in the lower bound and upper bound values is very low.<br> **Lower CI means lower degree of variability.**

### (a) Linear Train

**Here, the RSS for cubic regression is smaller than RSS for linear regression**<br>
The cubic model has more parameters, and hence it will fit the data at least as well as the linear model. Therefore RSS for cubic regression is smaller than linear regression.

### (b) Linear Test

**Here, the RSS for linear regression is smaller than RSS for cubic regression**<br>
The cubic model has more parameters. While training, it fits the known data, but while testing, it might cause overfitting as it might learn noise. Therefore, for test data, RSS for linear regression is smaller than cubic regression. 

### (c) Not Linear Train

**Here, the RSS for cubic regression is smaller than RSS for non-linear regression**<br>
A non-linear regression model can be too simple to capture actual pattern, which may lead to underfitting. Therefore, RSS for cubic regression is smaller in this case.

### (d) Not Linear Testing

**Here, we don't have enough information to tell**<br>
If the data is closer to linear, the linear model will likely have lower test RSS due to better generalization. If the relationship is highly non-linear, the cubic model may perform better by capturing complex patterns. Without knowing the true structure, we cannot determine whether increasing model complexity will improve or worsen test RSS.
