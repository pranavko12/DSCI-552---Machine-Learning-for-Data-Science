Import packages

```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, RocCurveDisplay
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

import warnings
warnings.filterwarnings('ignore')
```

### (a) Download the APS Failure data

```python
df_1 = '../data/aps_failure_training_set.csv'
df_2 = '../data/aps_failure_test_set.csv'
```

```python
df_train = pd.read_csv(df_1, skiprows=20, na_values = 'na')
df_test = pd.read_csv(df_2, skiprows=20, na_values = 'na')
df_train = df_train.replace({'neg': 0, 'pos': 1, 'na':np.nan})
df_test = df_test.replace({'neg': 0, 'pos': 1, 'na':np.nan})
```

```python
x_train = df_train.iloc[:, 1:]
y_train = df_train.iloc[:, 0]
```

```python
x_test = df_test.iloc[:, 1:]
y_test = df_test.iloc[:, 0]
```

### (b) Data Preparation

#### (i) Research what types of techniques are usually used

Usually, we use 3 techniques:
1. Delete the missing values:

    2. Pairwise Deletion
    
2. Imputation methods:
    1. Mean/Median/Mode Imputation

**Here, I am going to use Mean Imputation Method**

```python
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputed_x_train = imputer.fit_transform(x_train)
imputed_x_train = pd.DataFrame(imputed_x_train)
imputed_x_train.columns = x_train.columns

imputed_x_train
```

```python
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputed_x_test = imputer.fit_transform(x_test)
imputed_x_test = pd.DataFrame(imputed_x_test)
imputed_x_test.columns = x_test.columns

imputed_x_test
```

```python
combined_train = pd.concat([imputed_x_train, y_train], axis=1, join="inner")
combined_test = pd.concat([imputed_x_test, y_test], axis=1, join="inner")
combined_df = pd.concat([combined_train, combined_test])

combined_df
```

#### (ii) Calculate the coefficient of variation

```python
CV = {}
std = imputed_x_train.describe().loc['std']
mean = imputed_x_train.describe().loc['mean']
CV['CV'] = std/mean
CV_df = pd.DataFrame.from_dict(CV, orient='index')

CV_df.T
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(combined_df.corr())

plt.title("Correlation Matrix For Features")
plt.show()
```

```python
x_train_features = combined_df[cols]
sns.pairplot(x_train_features, diag_kind=None)
plt.title("Scatterplot for features")
plt.show()
```

```python
plt.figure(figsize=(12,8))
bp = sns.boxplot(data=combined_df[cols], orient = 'h')
bp.set(xscale = "log")
plt.title("Boxplot for features")
plt.show()
```

### (e) XGBoost and Model Trees

```python
y_train_xg = LabelEncoder().fit_transform(y_train.values.ravel())
y_test_xg = LabelEncoder().fit_transform(y_test.values.ravel())
```

```python
alphas = [10**i for i in range(-4, 4)]
mse = []

for a in alphas:
    model = XGBClassifier(verbosity=0, reg_alpha=a, use_label_encoder=False)
    accuracy = cross_val_score(estimator=model, X=imputed_x_train, y=y_train_xg, cv=10)
    print("alpha:", a)
    print("Cross validation score:", accuracy.mean())
    mse.append(accuracy.mean())
```

```python
best_alpha = alphas[np.argmax(mse)]
model = XGBClassifier(verbosity=0, reg_alpha=best_alpha)
model.fit(imputed_x_train, y_train_xg)
y_pred_test = model.predict(imputed_x_test)
test_accuracy = accuracy_score(y_test_xg, y_pred_test)
```

```python
print("Best Alpha (reg_alpha): ", best_alpha)
print("Best Cross-validation Accuracy: ", max(mse))
print("Test Accuracy:", test_accuracy)
```

```python
conf_matrix = confusion_matrix(y_train_xg, model.predict(imputed_x_train))
print("Confusion Matrix: \n", conf_matrix)
```

```python
roc_train = RocCurveDisplay.from_estimator(model, imputed_x_train, y_train)
plt.title("ROC XGBoost Train")
plt.show()
```

FOR TEST:

```python
conf_matrix = confusion_matrix(y_test_xg, model.predict(imputed_x_test))
print("Confusion Matrix: \n", conf_matrix)
```

```python
roc_test = RocCurveDisplay.from_estimator(model, imputed_x_test, y_test)
plt.title("ROC XGBoost Test")
plt.show()
```

**a. Steadily decreases.** 
   As s increses, the model gains more freedom to fit the train data. This means the training error will steadily decres=ase, as a more flexible model can always fit the training error better.

**b. Decreases initially, then starts to increase.**
At very low values of s, the modell is overly simplistic and underfits, so the test error is high. As s increases, the model stats to capture relevant patterns and test error decreases. 
There comes a point where model becomes too complex, causing overfitiing. This is where the test error increases.

**c. Steadily increases.** As s increases, we have more model complexity. More flexible mode is more sensity towards smal change. Hence, variance increases.

**d. Steadily decreases.** As s increases, variance increasees. Therefore, bias decreases.

**e. Its not affected.** Irreducible error is caused due to noise. It does not change with any change in s. It is out of our control and hence it is not affected.
